### 2.1 Audio Sampling
Audio sampling is the process of converting an audio signal into a digital representation by periodically measuring the amplitude of the audio signal. The number of samples per second is called the sample rate (sr), and a higher sample rate means that more samples are captured per second, resulting in better audio quality. The `librosa.load` function loads an audio file located at the specified path and returns the audio data as a numpy array and the sample rate as an integer.

```python
test_audio, sr = librosa.load(test_audio_path)

print(f"There are {test_audio.shape[0]} sampling points in this audio.")
print(f"The sampling rate is {sr}.")
```

We can use a combination of `librosa.display` and `matplotlib.pyplot` to present this test sample as a waveform plot in the time domain.

```python
import librosa.display
import matplotlib.pyplot as plt   

plt.figure(figsize=[15,2])   
librosa.display.waveshow(y=test_audio)
```

### 2.2 Log-Mel Feature Extraction
The Log-Mel spectrum is based on the Mel scale, which is a logarithmic scale that is used to represent the frequency of sound in a way that is more closely aligned with the human auditory system. The Mel spectrum is calculated by applying a nonlinear transformation to the power spectrum of a signal. The relationship between Mel frequency and Hz frequency can be defined as:

$$ Mel(f) = 2595\times \log_{10}(1 + \frac{f}{700}) = 1125\times \ln(1 + \frac{f}{700}) $$

where $f$ stands for the frequency of the signal in $Hz$. When performing SER, logarithmic operations are usually done on the Mel spectrum, which can compress the dynamic range of the Mel spectrum, making it easier to visualize and compare different spectra. The Log-Mel Spectrogram feature is currently a very commonly used feature in SER. Due to the powerful capabilities of CNN in processing images, the use of Log-Mel spectrogram features is becoming more and more extensive. We can use function `librosa.feature.melspectrogram()` to extract the Mel spectrum and use function `librosa.power_to_db()` to execute logarithmic operations.

```python
plt.figure(figsize=[15,30])

# Mel Spectrogram without Log transformation
mel_spectrogram = librosa.feature.melspectrogram(y=test_audio, sr=sr, n_fft=2048,hop_length=512,n_mels=64)
plt.subplot(1,2,1)
plt.title("Mel Spectrogram")
plt.imshow(mel_spectrogram)

# Log-Mel Spectrogram
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
plt.subplot(1,2,2)
plt.title("Log-Mel Spectrogram")
plt.imshow(log_mel_spectrogram)
```

### 2.3 Data Normalization
Data normalization is able to transform the values of features in the dataset to a common scale, which increases the generalization degree of the features. In the project, I use Z-score normalization which standardizes the values of a feature by using each value to subtract the mean of the feature data, and then dividing the result by the standard deviation of the feature as below:

$$x_{norm} = \frac{x - \frac{1}{n} \sum_{i=1}^{n} x_i}{\sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \frac{1}{n} \sum_{i=1}^{n} x_i)^2}}$$

We can use function `torch.mean()` to calculate the mean and use function `torch.std()` to calculate the standard deviation.

```python
import torch

t_log_mel_spectrogram = torch.tensor(log_mel_spectrogram) # Convert the numpy array into tensor
N_log_mel_spectrogram = (t_log_mel_spectrogram-torch.mean(t_log_mel_spectrogram))/ torch.std(t_log_mel_spectrogram)
print(f"The Log-Mel Spectrogram before normalization:\n\n {t_log_mel_spectrogram}.\n\n")
print(f"The Log-Mel Spectrogram after normalization:\n\n {N_log_mel_spectrogram}.")
```

### 2.4 Data Augmentation
In the project, I use `nlpaug` library to augment the data. `nlpaug` can be used to add noise to the spectrograms in a dataset, which includes time-domain additions and frequency-domain additions. Data augmentation are often used to increase the the size and diversity of the dataset. It is particularly useful in cases where there is less data, as it allows a wide range of variation in the input data to help improve the generalisation performance of the model.

```python
import nlpaug.augmenter.spectrogram as nas
import nlpaug.flow as naf
import random
```

`nas.FrequencyMaskingAug()` function is used to cover parts of the frequency domain, while `nas.TimeMaskingAug()` function is used to cover parts of the time domain. We can control the position and the width of the mask by changing the value of parameter `zone`.

```python
random_list=[] # Generate a random list.
for i in range(4):
    random_list.append(random.randint(0,1440))

img_num=1
plt.figure(figsize=[15,15])
for i in random_list:
    audio, sr = librosa.load(RAVDESS_df["Audio Path"][i]) # Randomly select 4 audio data in RAVDESS database.
    
    # Present the original Log-Mel spectrogram.
    original_mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_fft=2048,hop_length=512,n_mels=64)
    original_log_mel_spectrogram = librosa.power_to_db(original_mel_spectrogram)
    plt.subplot(4,2,img_num)
    plt.axis("off")
    plt.imshow(original_log_mel_spectrogram)
    plt.title("Original Spectrogram")
    
    # Present the augmented Log-Mel spectrogram.
    flow = naf.Sequential([
        nas.FrequencyMaskingAug(zone=(0, 1), factor=(4, 8), verbose=4),
        nas.TimeMaskingAug(zone=(0.5, 0.15))])
    aug_data = flow.augment(original_log_mel_spectrogram)
    plt.subplot(4,2,img_num+1)
    plt.axis("off")
    plt.imshow(aug_data[0])
    plt.title("Augmented Spectrogram")
    
    img_num+=2
```

### 2.5 Spectrogram Padding
The data in the speech database has different durations and when converted into a spectrogram will appear to be unequal in width. Forcing them to be resized to equal lengths would lose some of the time-related information, so I select the method of padding these spectrograms. Spectrogram padding is the process of adding zeros to the edges of a spectrogram in order to increase its size. This is typically done to ensure that the input data to the model has a consistent size. We can call the `pad()` function by importing `torch.nn.functional`.

```python
import torch.nn.functional as F
```

The length of the padding depends on the longest data in the database. The length of the padding depends on the longest data in the database. In RAVDESS, the longest data length is 228. To facilitate multiple pooling, I padding all the spectrograms to the length of 240. Here is a demonstration of the padding effect for the test audio.

```python
plt.figure(figsize=[15,8])

# Padding step
padding_log_mel_spectrogram= F.pad(N_log_mel_spectrogram, (0,240-N_log_mel_spectrogram.shape[1],0,0))

plt.subplot(1,2,1)
plt.title("Original Log-Mel Spectrogram")
plt.imshow(N_log_mel_spectrogram)

plt.subplot(1,2,2)
plt.title("Padding Log-Mel Spectrogram")
plt.imshow(padding_log_mel_spectrogram)
```

### 4.5 TinyVGG Model
The architecture of TinyVGG typically consists of several convolutional layers , activation layers and pooling layers, followed by a few fully connected layers, and finally an output layer for classifying the input tensor into one of several predefined classes. Each of the layer performs a specific function in the process of transforming the input into a set of output features:<br>• Input layer: Receiving the input image or tensor.<br>• Convolutional layer: Performing the convolution operation on the input data.<br>• Activation layer: Applying a non-linear function to the output of the convolutional layer.<br> • Pooling layer: Reducing the dimensionality of the feature maps produced by the convolutional layer.<br> • Fully connected layer: Receiving the output of the previous layers and computing a set of weighted sum outputs.<br>• Output layer: producing the final output of the network.<br>
<img src="file_img/network.png" width="900" height="600">

In this project, I set up three blocks in the `TinyVGG` model, performed six convolutions and three poolings. The architecture of this model can be viewed in file `TinyVGG.py`.

```python
import TinyVGG
importlib.reload(TinyVGG)
```

I put the `train_step` function and the `val_step` function into the module `TrainAndVal`.

```python
import TrainAndVal
importlib.reload(TrainAndVal)
```

```python
from torchmetrics import Accuracy
import torch.nn as nn
from tqdm.auto import tqdm

torch.cuda.manual_seed(30)

epochs = 100
checkpoint_frequency=3

model_TinyVGG = TinyVGG.TinyVGG_Model(input_shape=1,hidden_units=10,output_shape=8).to(DEVICE)
acc_fn = Accuracy(num_classes=8).to(DEVICE)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(params=model_TinyVGG.parameters(),lr=0.01)
schd = ExponentialLR(optimizer,gamma=1)
stopper =  EarlyStopper(patience=10, tolerance=0.2)
writer = SummaryWriter()

for epoch in tqdm(range(epochs)):
  
    print(f"Epoch: {epoch}\n-------")
    
    TrainAndVal.train_step(epoch=epoch,
               model=model_TinyVGG,
               data_loader=RAVDESS_train_dl, 
               loss_fn=loss_fn, 
               optimizer=optimizer,
               lr_scheduler=schd,
               acc_fn = acc_fn,
               writer = writer,
               DEVICE = DEVICE
              )
    
    if (epoch % checkpoint_frequency == 0):
        Checkpointing.TinyVGG_save_checkpoint(model=model_TinyVGG,epoch=epoch, save_dir="./saved_models" )
    
    if (TrainAndVal.val_step(epoch=epoch, model=model_TinyVGG,data_loader=RAVDESS_test_dl, loss_fn=loss_fn, 
        early_stopper=stopper, acc_fn = acc_fn, writer = writer, DEVICE = DEVICE)==0):
        break
```

### 4.6 ResNet Model
As we can see from the TinyVGG training process above, the TinyVGG model is very prone to gradient disappearance, which is a common problem with CNN models. To address this problem, the researchers proposed the ResNet (Residual Network) model, whose innovation is the use of residual connectivity, which allows the network to bypass one or more layers and propagate the input directly to deeper layers. This allows the network to go much deeper than a traditional CNN without suffering from the gradient disappearance problem. The architecture of ResNet model can be viewed in file `ResNet.py`.

<img src="file_img/Resnet.png" width="900" height="600">

```python
import ResNet
importlib.reload(ResNet)
```

```python
from torchmetrics import Accuracy
import torch.nn as nn
from tqdm.auto import tqdm

torch.cuda.manual_seed(30)

epochs = 50
checkpoint_frequency=3

model_ResNet = ResNet.get_resnet().to(DEVICE)
acc_fn = Accuracy(num_classes=8).to(DEVICE)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(params=model_ResNet.parameters(),lr=0.01)
schd = ExponentialLR(optimizer,gamma=1)
stopper =  EarlyStopper(patience=5, tolerance=0.1)
writer = SummaryWriter()


for epoch in tqdm(range(epochs)):
    print(f"Epoch: {epoch}\n-------")
    
    TrainAndVal.train_step(epoch=epoch,
               model=model_ResNet,
               data_loader=RAVDESS_train_dl, 
               loss_fn=loss_fn, 
               optimizer=optimizer,
               lr_scheduler=schd,
               acc_fn = acc_fn,
               writer = writer,
               DEVICE = DEVICE
              )
    
    if (epoch % checkpoint_frequency == 0):
        Checkpointing.ResNet_save_checkpoint(model=model_ResNet,epoch=epoch, save_dir="./saved_models" )
    
    if (TrainAndVal.val_step(epoch=epoch, model=model_ResNet,data_loader=RAVDESS_test_dl, loss_fn=loss_fn, 
        early_stopper=stopper, acc_fn = acc_fn, writer = writer, DEVICE = DEVICE)==0):
        break
```
